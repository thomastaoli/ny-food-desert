```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
# Request the following URL
url = 'https://locations.traderjoes.com/ny/new-york/'
response = requests.get(url)

# Parse the HTML of the page
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
addresses = []

# Find all divs with class 'address-left'
for div in soup.find_all('div', class_='address-left'):
    address = {}
    spans = div.find_all('span')
    
    # Assuming the street address is always the second span
    address['street'] = spans[1].text.strip()
    
    # Assuming the zip code is always the second to last span
    address['zip_code'] = spans[-3].text.strip()
    
    addresses.append(address)

# Print the addresses
for address in addresses:
    print(f"Street: {address['street']}, Zip: {address['zip_code']}")
```
